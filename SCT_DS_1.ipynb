# Selecting the relevant columns

```python
df_population = df[['Country Name', '2022']]
```

# Sorting by population and selecting the top 10 countries for better visualization

```python
df_top10 = df_population.sort_values(by='2022', ascending=False).head(10)
```

```python
print(df_top10)
```

# Plotting the bar chart

```python
plt.figure(figsize=(12, 6))
plt.barh(df_top10['Country Name'], df_top10['2022'], color='skyblue')
plt.xlabel('Total Population in 2022')
plt.ylabel('Country')
plt.title('Top 10 Most Populated Countries in 2022')
plt.gca().invert_yaxis()
plt.show()
```

# Histogram of Population Distribution

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df_population['2022'], bins=30, kde=True, color='purple')
plt.xlabel('Population in 2022')
plt.ylabel('Frequency')
plt.title('Distribution of Population in 2022')
plt.show()
```

# Pie Chart of Top 10 Populated Countries

```python
plt.figure(figsize=(10, 6))
plt.pie(df_top10['2022'], labels=df_top10['Country Name'], autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Population Share of Top 10 Countries in 2022')
plt.show()
```
